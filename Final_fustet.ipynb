# Data processing

## Data 1

### Data1 clean1

```python

TF1=list() ## กำหนด TF ให้เป็น list ว่าง เพื่อใช้เติมค่า loop
for each_row in data1.iterrows(): ## loop each_row ใน data1
    if len(str(each_row[1]['user']))== 5 : #สร้างเงื่อนไข user ที่มีจำนวนตัวอักษรเท่ากับ 5
        if str(each_row[1]['user']) <= '99999' : #หาก user มีค่า น้อยกว่า 100000
            TF1.append(True) ## หากเงื่อนไขถูกต้องให้เพิ่มเข้าไปใน TF
        elif str(each_row[1]['user']) >= '99999' : #หาก user มีค่า มากกว่า 100000
            TF1.append(False) ## หากเงื่อนไขไม่ถูกต้อง ไม่ให้เพิ่มเข้าไปใน TF
    else:
        TF1.append(False) ## หากเงื่อนไขไม่ถูกต้อง ไม่ให้เพิ่มเข้าไปใน TF

data1[TF1] ## แสดงผล
```

```python
df1 =data1[TF1]
```

```python
df1_clean = df1[df1['user']!= '6724-']  ##พบค่าผิดปกติ คือ 6724- จึงทำการตัดออก
```

### Data 1 clean2

```python
TF2=list() ## กำหนด TF ให้เป็น list ว่าง เพื่อใช้เติมค่า loop
for each_row in df1_clean.iterrows(): ## loop each_row ใน data1
    if len(str(each_row[1]['site']))== 5 : #สร้างเงื่อนไข user ที่มีจำนวนตัวอักษรเท่ากับ 5
        if str(each_row[1]['site']) <= '99999' : #หาก user มีค่า น้อยกว่า 100000
            TF2.append(True) ## หากเงื่อนไขถูกต้องให้เพิ่มเข้าไปใน TF
        elif str(each_row[1]['site']) >= '99999' : #หาก user มีค่า มากกว่า 100000
            TF2.append(False) ## หากเงื่อนไขไม่ถูกต้อง ไม่ให้เพิ่มเข้าไปใน TF
    else:
        TF2.append(False) ## หากเงื่อนไขไม่ถูกต้อง ไม่ให้เพิ่มเข้าไปใน TF

df1_clean[TF2] ## แสดงผล
```

```python
df1_clean2 = df1_clean[TF2]
```

### Data 1 clean 3

```python
TF3=list() ## กำหนด TF ให้เป็น list ว่าง เพื่อใช้เติมค่า loop
for each_row in df1_clean2.iterrows(): ## loop each_row ใน data1
    if len(str(each_row[1]['PartitionKey']))== 5 : #สร้างเงื่อนไข user ที่มีจำนวนตัวอักษรเท่ากับ 5
        if str(each_row[1]['PartitionKey']) <= '99999' :  #หาก user มีค่า น้อยกว่า 100000
            TF3.append(True) ## หากเงื่อนไขถูกต้องให้เพิ่มเข้าไปใน TF
        elif str(each_row[1]['PartitionKey']) >= '99999' : #หาก user มีค่า มากกว่า 100000
            TF3.append(False) ## หากเงื่อนไขไม่ถูกต้อง ไม่ให้เพิ่มเข้าไปใน TF
    else:
        TF3.append(False) ## หากเงื่อนไขไม่ถูกต้อง ไม่ให้เพิ่มเข้าไปใน TF

df1_clean2[TF3] ## แสดงผล
```

# Change dtype

##Data 1

```python
data1_use.info() ##เช็คข้อมูล dtype แต่ละคอลัมน์
```

```python
data1_use['time'] = pd.to_datetime(data1_use['time'],format='%H:%M%:%S',errors='ignore') ##แก้ dtype คอลัมน์ time
```

```python
data1_use['date'] = pd.to_datetime(data1_use['date'],format='%Y/%m/%d') ##แก้ dtype คอลัมน์ date
```

```python
data1_use['phone']=data1_use['phone'].astype('string') ##แก้ dtype คอลัมน์ phone
```

```python
data1_use['urine']=data1_use['urine'].astype('string') ##แก้ dtype คอลัมน์ urine
```

```python
data1_use['RowKey']=data1_use['RowKey'].astype('string') ##แก้ dtype คอลัมน์ RowKey
```

```python
data1_use['user']=data1_use['user'].astype('int64') ##แก้ dtype คอลัมน์ user
```

```python
data1_use['site']=data1_use['site'].astype('int64') ##แก้ dtype คอลัมน์ site
```

```python
data1_use['PartitionKey']=data1_use['PartitionKey'].astype('int64') ##แก้ dtype คอลัมน์ PartitionKey
```

```python
data1_use.info() ## เช็คข้อมูลแต่ละคอลัมน์
```

```python
data1_use ## แสดงตาราง
```

## Data 2 

```python
data2_use.info() ## เช็คข้อมูลแต่ละคอลัมน์
```

```python
data2_use['hcode']=data2_use['hcode'].astype('int64') ## แก้ dtype คอลัมน์ hcode
```

# Merge Table

```python
df_merge = data1_use.merge(data2_use,how='left',left_on='site',right_on='hcode') ## เชื่อมตาราง data1_use กับ data2_use ด้วยคีย์ site และ hcode
df_merge  ## แสดงตาราง
```

# Export

```python
data_processing=df_merge[['user','hcode','name','province','amphur','date','time','ov','phone','urine']]  ##เลือกคอลัมน์ที่ต้องการ
data_processing  ## แสดงตาราง
```

```python
data_processing.shape  ##เช็คขนาดของตาราง
```

```python
data_complete = data_processing.dropna() ##กำจัด missing
data_complete.shape ##เช็คขนาดของตาราง
```
