## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

###HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

````python
ad = list() #กำหนด ad รับค่า list ว่าง
for each_row in data_covid.iterrows(): #ใช้คำสั่ง .iterrows() เข้าถึงข้อมูล data_covid ทีละแถว วนลูปซ้ำไปเรื่อยๆ 
  if int(each_row[1]['announce_date'].split('/')[1]) <=20: #เอาข้อมูลตั้งแต่แถวที่ 1 คอลัมน์ announce_date ใช้ .spilt('/') เพื่อแยกข้อมูลออก ใส่เลข 1 เพื่อเอาตำแหน่งที่ 1 ใช้ int เพื่อแปลงเป็นตัวเลข แล้วเอาตัวเลข <=20
    ad.append(True) #ถ้าจริง เพิ่มสมาชิกเข้าไป
  else:
    ad.append(False) #ถ้าไม่จริง ไม่เพิ่มสมาชิกเข้าไป
data_covid[ad] #แสดงผล
````
